```python
list_of_addresses1 = df1['Full Address']
list_of_addresses2 = df1_original['modified_clean_address']
```

```python
ratios_list1 = {}
for i in range(0,len(list_of_addresses2)):
    #if list_of_addresses2[i] != list_of_addresses1[i]:
    #print(i, list_of_addresses1[i], list_of_addresses2[i])
    #print(i, df1['Full Name'][i], list_of_addresses2[i])
    try:
        ratio1 = fuzz.ratio(list_of_addresses1[i].lower(), list_of_addresses2[i].lower())
        #print(ratio1,'%')
        ratios_list1[i] = ratio1
    except:
        ratios_list1[i] = -1
    try:
        ratio2 = fuzz.ratio(df1['Full Name'][i].lower(), list_of_addresses2[i].lower())
        #print(ratio2,'%')
        try:
            if ratio2 > ratios_list1[i]:
                #print('in loop2', ratio1, ratio2)
                ratios_list1[i] = ratio2
        except:
            pass
    except:
        pass
```

```python
df1_original['ratio'] = ratios_list1.values()
```

```python
df1_original.sort_values(by='ratio', ascending=True)
```
