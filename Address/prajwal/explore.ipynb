```python
df['filter1'] = np.where(df['Occupation'].isna() & df['H House Number'].isna() & df['new h address-flag'].isna(), 1, 0)
```

```python
df['filter2'] = np.where(~df['H House Number'].isna() & df['H Street Name'].isna() &df['Address 2 Street Name'].isna(), 1, 0)
```

```python
df['filter3'] = np.where((df['Full Address'].str.find('.5')>0) & df['H House Number'].isna(),1,0)
```

```python
df['filter4'] = np.where(df['sus']==2.0, 1,0)
```

```python
df['filter5'] = np.where((df['Full Address'].str.find('.5')>0) & ~df['H House Number'].isna() & (df['H Status-flag']=='Assume'), 1,0)

```

```python
df['filter6'] = np.where((df['H Status-flag']=='Assume')&(df['Full Address'].str.find(' H ')>0),1,0)
```

```python
df['filter8'] = np.where((df['H Status-flag']=='Assume')&(df['Full Address'].str.find(' h ')>0)&(df['new h address-flag'].isna()),1,0)

```
